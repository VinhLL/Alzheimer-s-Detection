```python
import os
import pandas as pd 
import re
```

```python
df = pd.read_csv('data/dataset.csv')
df.head()
```

```python
df['mri_link'] = df['mri_link'].str.replace(' ', '_')
df.head()
```

```python
df['mri_link'][0]
```

```python
df.count()
```

```python
duplicate_row = df[df.duplicated(keep=False)]
duplicate_row
```

```python
data_dir = 'data/MRI'

full_df = pd.DataFrame(columns=['subject_id', 'mri_description', 'mri_time', 'image_id'])

for f1 in os.listdir(data_dir):
    if os.path.isdir(os.path.join(data_dir, f1)):
        for f2 in os.listdir(os.path.join(data_dir, f1)):
            if os.path.isdir(os.path.join(data_dir, f1, f2)):
                for f3 in os.listdir(os.path.join(data_dir, f1, f2)):
                    if os.path.isdir(os.path.join(data_dir, f1, f2, f3)):
                        for f4 in os.listdir(os.path.join(data_dir, f1, f2, f3)):
                            if os.path.isdir(os.path.join(data_dir, f1, f2, f3, f4)):
                                full_df = pd.concat([full_df, pd.DataFrame({'subject_id': [f1], 'mri_description': [f2], 'mri_time': [f3], 'image_id': [f4]})], ignore_index=True)

full_df.head()
```

```python
full_df['mri_link'] = 'data/MRI/' + full_df['subject_id'] + '/' + full_df['mri_description'] + '/' + full_df['mri_time'] + '/' + full_df['image_id']

full_df['mri_link'][0]
```

```python
df[['subject_id', 'mri_description', 'image_id']] = df['mri_link'].str.extract(r'MRI/([^/]+)/([^/]+)/[^/]+/([^/]+)$')
df['mri_description'] = df['mri_description'].str.replace(r'[\(\)]', '_', regex=True)
df['mri_description'] = df['mri_description'].str.replace('>', '_', regex=False)

df.head()
```

```python
df['key'] = df['subject_id'] + '|' + df['mri_description'] + '|' + df['image_id']
full_df['key'] = full_df['subject_id'] + '|' + full_df['mri_description'] + '|' + full_df['image_id']

link_lookup = full_df[['key', 'mri_link']].rename(columns={'mri_link': 'mri_link_2'})

df = df.merge(link_lookup, on='key', how='left')
df.drop(columns=['key'], inplace=True)

df['mri_link_2'] = df['mri_link_2'].astype(str)
```

```python
df['mri_exists'] = df['mri_link_2'].apply(lambda path: os.path.exists(path))

missing_mri = df[~df['mri_exists']]
missing_mri
```

```python
df['mri_link'] = df['mri_link_2']
df = df.drop(columns=['subject_id', 'mri_description', 'image_id', 'mri_exists', 'mri_link_2'])
df
```

```python
df.to_csv('final_dataset.csv')
```
