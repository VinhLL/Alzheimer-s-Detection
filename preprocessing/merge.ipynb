```python
import os
import pandas as pd
```

```python
dti_df = pd.read_csv('data\dti.csv')
dti_df.head()
```

```python
dti_df.info()
```

```python
mri_df = pd.read_csv('data\mri.csv')
mri_df.head()
```

```python
mri_df.info()
```

```python
df = pd.merge(dti_df, mri_df, on=['subject_id', 'DIAGNOSIS'], how='inner')
df.head()
```

```python
df.info()
```

```python
# 1 cột mới ở df là số ngày giữa dti_date và mri_date
df['dti_date'] = pd.to_datetime(df['dti_date'])
df['mri_date'] = pd.to_datetime(df['mri_date'])

df['days_between'] = abs((df['mri_date'] - df['dti_date']).dt.days)
df.head()
```

```python
df['days_between'].unique()
```

```python
df = df[df['days_between'] < 7]
df.head()
```

```python
df['days_between'].unique()
```

```python
df.info()
```

```python
# in những dòng bị trùng subject_id

df1 = df[df.duplicated(subset=['dti_link'], keep=False)].sort_values(by='subject_id')
```

```python
# xóa cột khong cần thiết
df = df.drop(columns=['dti_date', 'mri_date', 'days_between', 'subject_id'])
```
